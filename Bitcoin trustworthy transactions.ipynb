Q7. Find the influential nodes in the network using different centrality: Degree centrality; Closeness centrality; Betweenness centrality; Eigenvector centrality. Present any one which gives you relevant results. 

```python
deg_centrality = nx.degree_centrality(E)
print("Degree Centrality is = ", deg_centrality)

close_centrality = nx.closeness_centrality(E)
print("Closeness Centrality is = ", close_centrality)

bet_centrality = nx.betweenness_centrality(E)
print("Betweenness Centrality is = ", bet_centrality)

eigenvector_centrality = nx.eigenvector_centrality(E)
print("Eigenvectors Centrality is = ", eigenvector_centrality)
```

```python
df1 = pd.DataFrame.from_dict([deg_centrality, close_centrality, bet_centrality, eigenvector_centrality])
df1 = pd.DataFrame.transpose(df1)
df1.columns = ['DEGREE_CENTRALITY', 'CLOSENESS_CENTRALITY', 'BETWEENNESS_CENTRALITY', 'EIGENVECTOR_CENTRALITY']
```

```python
#Sorting by Degree_CENTRALITY
df1.sort_values('DEGREE_CENTRALITY', ascending = False)[:10]
```

19. Use Two star recommendation algorithm in your project.

```python
recommended = defaultdict(int)

for n, d in E.nodes(data=True):

    for n1, n2 in combinations(E.neighbors(n), 2):

        if not E.has_edge(n1, n2):

            recommended[((n1), (n2))] += 1
```
